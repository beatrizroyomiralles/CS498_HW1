```python
import pandas as pd
import os

# Ver los archivos en el directorio de entrada
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# Cargar el dataset con el nombre de archivo corregido
df = pd.read_csv('/kaggle/input/youtube-video-and-channel-analytics/YouTubeDataset_withChannelElapsed.csv')

# Revisar las primeras filas del dataset
df.head()

# Verificar los nombres de las columnas para identificar los correctos
print(df.columns)

# Asegurarse de que cada videoId sea único
# Usar los nombres de columnas correctos según el dataset
df = df.groupby('videoId').agg({
    'videoLikeCount': 'sum',
    'videoDislikeCount': 'sum',
    'VideoCommentCount': 'sum',
    'videoViewCount': 'sum'
}).reset_index()

# 1. Calcular el promedio de likes por video
average_likes = df['videoLikeCount'].mean()
print(round(average_likes))

# 2. Tasa de participación (engagement rate)
df['engagement_rate'] = (df['videoLikeCount'] + df['videoDislikeCount'] + df['VideoCommentCount']) / df['videoViewCount']
average_engagement_rate = df['engagement_rate'].mean()
print(round(average_engagement_rate, 4))

# 3. Tamaño de almacenamiento por video
metadata_size = 500  # bytes
thumbnail_size = 200 * 1024  # 200 KB en bytes
total_size_mb = (metadata_size + thumbnail_size) / (1024 * 1024)
print(round(total_size_mb, 4))

# 4. Transferencia de datos total por interacciones
total_engagements = df['videoLikeCount'].sum() + df['videoDislikeCount'].sum() + df['VideoCommentCount'].sum()
total_data_transfer_mb = total_engagements * total_size_mb
print(round(total_data_transfer_mb, 4))

# 5. Estimar vistas por minuto (asumiendo 5 años)
total_views = df['videoViewCount'].sum()
minutes_in_5_years = 5 * 365 * 24 * 60
views_per_minute = total_views / minutes_in_5_years
print(round(views_per_minute))

# 6. Datos accedidos por minuto para mostrar detalles de video
data_per_minute = views_per_minute * total_size_mb
print(round(data_per_minute, 4))
```
